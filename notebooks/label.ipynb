```python
import os
import shutil
from random import sample

# Define the source directory where your class folders are
source_dir = '../data/images/dermnet/test1/'

# Define the destination directory where you want to copy the selected images
destination_dir = '../data/images/dermnet/test_combined/'

# Make sure the destination directory exists
os.makedirs(destination_dir, exist_ok=True)

# Define the classes and the number of images you want to select from each
classes = {
    'Atopic Dermatitis Photos': 32,
    'Seborrheic Keratoses and other Benign Tumors': 8,
    'Lupus and other Connective Tissue diseases': 17,
    'Scabies Lyme Disease and other Infestations and Bites': 25
}

# Iterate over each class
for class_name, num_images in classes.items():
    # Define the path to the class folder
    class_folder = os.path.join(source_dir, class_name)
    destf = os.path.join(destination_dir,class_name)
    
    # Get all the image files in the class folder
    all_images = [f for f in os.listdir(class_folder) if os.path.isfile(os.path.join(class_folder, f))]
    
    # Randomly select the specified number of images
    selected_images = sample(all_images, num_images)
    
    # Copy the selected images to the destination folder
    for image in selected_images:
        # Define the source image path
        src_image_path = os.path.join(class_folder, image)
        
        # Define the destination image path
        dest_image_path = os.path.join(destf, image)
        
        # Copy the image to the destination folder
        shutil.copy2(src_image_path, dest_image_path)

print('Images have been copied.')
```

```python
df = pd.read_csv("../data/symptoms4.csv")
```

```python
df['Disease'] = df['Disease'].replace('Benign Tumors', 'Seborrheic Keratoses and other Benign Tumors')
```

```python
df.groupby(df['Disease']).count()
```

```python
df.head()
df = df[['Disease','Symptoms']]
```

```python
df_with_symptoms = df[df['Symptoms'].notna()]
```

```python
len(df_with_symptoms)
```

```python
sortedd = df_with_symptoms.sort_values("Disease")
```

```python
df_with_symptoms["Disease"].unique()
```

```python
sortedd = sortedd.sort_values("Disease")
```

```python
sortedd
```

```python

# Assuming 'sortedd' is your DataFrame and 'Disease' is the column with the disease names

# Define the counts for the training set for each disease
train_counts = {
    "Atopic Dermatitis": 128,
    "Lupus and other Connective Tissue diseases": 68,
    "Scabies Lyme Diease and other Infestations and Bites": 102,
    "Seborrheic Keratoses and other Benign Tumors": 33,
}

# Create empty DataFrames for the train and test sets
train_symp = pd.DataFrame()
test_symp = pd.DataFrame()

# Loop through the diseases and their counts
for disease, count in train_counts.items():
    # Filter the dataset for the disease
    disease_df = sortedd[sortedd['Disease'] == disease]
    
    # Sample the specified number of rows for the training set
    train_symp_disease = disease_df.iloc[:count]
    
    # The remaining rows will be part of the test set
    test_symp_disease = disease_df.iloc[count:]
    
    # Append these to the train and test DataFrames
    train_symp = pd.concat([train_symp, train_symp_disease])
    test_symp = pd.concat([test_symp, test_symp_disease])

# Now you have 'train_symp' and 'test_symp' as per your specified counts
```
