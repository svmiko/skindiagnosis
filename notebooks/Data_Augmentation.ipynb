```python
# Function to rotate images in a directory and save them
def rotate_and_save(input_dir, output_dir, angle=45):
    for class_folder in os.listdir(input_dir):
        class_input_dir = os.path.join(input_dir, class_folder)
        class_output_dir = os.path.join(output_directory, class_folder)
        os.makedirs(class_output_dir, exist_ok=True)  # Create class-specific output folder

        for filename in os.listdir(class_input_dir):
            if filename.endswith('.jpg') or filename.endswith('.png'):
                input_path = os.path.join(class_input_dir, filename)
                img = cv2.imread(input_path)
                # Rotate the image by the specified angle
                rotated_img = cv2.warpAffine(img, cv2.getRotationMatrix2D((img.shape[1] / 2, img.shape[0] / 2), angle, 1.0), (img.shape[1], img.shape[0]))
                output_path = os.path.join(class_output_dir, f"rotated_{angle}_{filename}")
                cv2.imwrite(output_path, rotated_img)
                print(f"Rotated {filename} and saved as {output_path}")

# Rotate images in the input directory and save them to class-specific output directories
rotate_and_save(input_directory, output_directory, angle=45)
```

```python
# Initialize the ImageDataGenerator with augmentation parameters
from tensorflow.keras.preprocessing.image import ImageDataGenerator1
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.8, 1.2]
)
```

```python
# Function to perform data augmentation and save augmented images
def augment_data(input_dir, output_dir, num_augmented_images=10):
    for class_folder in os.listdir(input_dir):
        class_input_dir = os.path.join(input_dir, class_folder)
        class_output_dir = os.path.join(output_directory, class_folder)
        os.makedirs(class_output_dir, exist_ok=True)  # Create class-specific output folder

        for filename in os.listdir(class_input_dir):
            if filename.endswith('.jpg') or filename.endswith('.png'):
                input_path = os.path.join(class_input_dir, filename)
                img = cv2.imread(input_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert image to RGB format
                img = img.reshape((1,) + img.shape)  # Reshape for flow() method

                # Generate augmented images
                i = 0
                for batch in datagen.flow(img, batch_size=1, save_to_dir=class_output_dir, save_prefix='augmented', save_format='jpeg'):
                    i += 1
                    if i >= num_augmented_images:
                        break  # Break the loop after generating the desired number of augmented images
```

```python
 augment_data(input_directory, output_directory, num_augmented_images=10)
```
